```python
import pandas as pd
from sklearn import tree
import pickle
```

```python
df=pd.read_csv('forest_fire.csv')
df
```

```python
df=df.drop(['Oxygen'],axis=1)
df
```

```python
df=df.drop(['Area'],axis=1)
df
```

```python
df.isnull().values.any()
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,2]
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
x_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=50) # with n_estimatiors=(500)tree number
rfc.fit(x_train,y_train)
rfc.score(x_test,y_test)
```

```python
predict=rfc.predict([[20,70]])
predict
```

```python
ans=rfc.predict_proba([[20,70]])
```

```python
print(ans[0])
```

```python
a=ans.max()
if(predict[0]<1):
    print(f"Predicted fire possibility= {100-(int(a*100))} & Unfire Possibility {int(a*100)} %")
else:
    print(f"Predicted fire possibility= {int(a*100)} % & Unfire Possibility = {100-(int(a*100))} %")
```

```python
pickle.dump(rfc,open("rfc.pkl","wb"))
```
